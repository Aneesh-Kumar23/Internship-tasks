# **Introduction and Problem Statement:**
The objective of this task is to explore and visualize the Iris dataset using Python libraries such as Pandas, Matplotlib, and Seaborn. The task focuses on understanding the dataset structure, summarizing the data, and creating visualizations to identify patterns and relationships among variables.

```python
#import libraries.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#load DataSet.
iris = sns.load_dataset('iris')
iris

#Dataset Understanding and Description.
print("Shape:", iris.shape)

print("Columns:")
print(iris.columns)

iris.head()
iris.info()
iris.describe()

#Data Cleaning and Preparation.
iris.isnull().sum()
print("No missing values found.")
print("Duplicate Rows:", iris.duplicated().sum())
iris = iris.drop_duplicates()

#Exploratory Data Analysis (EDA).
iris['species'].value_counts()
sns.pairplot(iris, hue='species')
plt.show()
```

### **Scatter Plot:**

```python
#Scatter Plot.
plt.figure(figsize=(8,6))

sns.scatterplot(
    x='sepal_length',
    y='petal_length',
    hue='species',
    data=iris
)
plt.title("Sepal Length vs Petal Length")
plt.show()
```

### **Scatter Plot Observation:**
The scatter plot shows a clear relationship between sepal length and petal length. Setosa is clearly separated from the other species.

### **Histogram:**

```python
#Histogram.
plt.figure(figsize=(8,6))

sns.histplot(
    iris['sepal_length'],
    bins=10,
    kde=True
)
plt.title("Distribution of Sepal Length")
plt.show()
```

### **Histogram Observation:**
Most flowers have sepal lengths between 5 cm and 7 cm.

### **Box Plot:**

```python
#Box Plot.
plt.figure(figsize=(8,6))

sns.boxplot(
    x='species',
    y='petal_length',
    data=iris
)
plt.title("Petal Length by Species")
plt.show()
```

### **Box Plot Observation:** 
The box plot shows variation among species and helps identify outliers.
